```python
print("The number of days=",{(data.Date.max()-data.Date.min()).days})
```

```python
x=data.drop(['Close','Date'],axis=1)
```

```python
y=data['Close']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
data_predict=rf.predict(x_test)
```

```python
print("The score for the prediction",r2_score(data_predict,y_test))
```

```python
input_list=[19,25,22,21,10000]
input_array=np.asarray(input_list)
input_array_reshape=input_array.reshape(1,-1)
yes=rf.predict(input_array_reshape)
print(yes)
```
