```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score
```

```python
house_data=sklearn.datasets.load_boston()
```

```python
print(house_data)
```

```python
house_dataframe=pd.DataFrame(house_data.data,columns=house_data.feature_names)
```

```python
house_dataframe.head(10)
```

```python
house_dataframe['price']=house_data.target
```

```python
x=house_dataframe.drop('price',axis=1)
y=house_dataframe['price']
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
pred_data=rf.predict(x_test)
```

```python
print("the accuracy is",r2_score(pred_data,y_test))
```

```python
plt.plot(y_test,pred_data)
plt.xlabel("predicted data")
plt.ylabel("test data")
plt.title("pred data vs actual data")
plt.show()
```
